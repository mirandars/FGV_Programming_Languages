#### 1) Construa uma função que retorne o enésimo número primo

```r
lista.primos <- c(2)
gera.primos <- function(n) 
{
    if(n == 1){
        return(2)
    } else {
        i <-3
        repeat {
            e_primo <- TRUE
            for (p in lista.primos) {
                if(i%%p == 0){
                    e_primo <- FALSE
                }
            }
            if(e_primo == TRUE){
                lista.primos <- append(lista.primos, i)
            }
            i <- i + 1
            if(length(lista.primos) == n){
                return(tail(lista.primos, n=1))
            }
        }
    }
}

for (numero in 1:10) {
    print(gera.primos(numero))
    }
```

```r
crivo.eratostenes <- function(n)
{
   n <- as.integer(n)
   if(n > 1e6) stop("n too large")
   primes <- rep(TRUE, n)
   primes[1] <- FALSE
   last.prime <- 2L
   for(i in last.prime:floor(sqrt(n)))
   {
      primes[seq.int(2L*last.prime, n, last.prime)] <- FALSE
      last.prime <- last.prime + min(which(primes[(last.prime+1):n]))
   }
   which(primes)
}

 crivo.eratostenes(1000)
```

#### 2) Construa uma função para gerar os n primeiros números da sequência de fibonacci

```r
fib <- function(n)
{
    fibvals <- numeric(n)
    fibvals[1] <- 1
    fibvals[2] <- 1
    for (i in 3:n) {
        fibvals[i] <- fibvals[i-1]+fibvals[i-2]
        }
    print(fibvals)
}

fib(10)
```

#### 3) Construa uma função para retornar a sequência de Collatz de um número n

```r
n <- 27
n.total <- NULL
while(n != 1){
  n <- ifelse(n%%2==0, n/2, 3*n +1)
  n.total <- c(n.total,n)
}

print(n.total)
```
